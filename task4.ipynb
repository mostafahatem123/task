```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
data=pd.read_csv(r"C:\Users\Mega Tech\OneDrive\Desktop\amazon_sales_dataset.csv")
data.head(5)
```

```python
encoder=OneHotEncoder(sparse_output=False)
one=encoder.fit_transform(data[['customer_region']].head(12))
```

```python
one
```

```python
data['customer_region']=data['customer_region'].astype(str)
```

```python
feature_names = encoder.get_feature_names_out(['customer_region'])
feature_names
```

```python
data_encoder=pd.DataFrame(one,columns=feature_names)
```

```python
data_encoder
```

```python
ind=data.columns.get_loc('customer_region')
ind
```

```python
before = data.iloc[:, :ind]
after = data.iloc[:, ind+1:]

data_final = pd.concat([before, data_encoder, after], axis=1)
data_final.head(10)
```
